# Replace

```python
import pandas as pd
```

```python
df=pd.read_csv(r"D:\dataset for Machine Learning\Fortune_10.csv") 
df
```

```python
df.replace(to_replace='health',value='finance') # to_replace replace any value of dataframe with value contain by value=""
```

```python
df.replace(10,15)     # shortcut of the to_replace
```

```python
df.replace([1,2,3,4,5],77) # you can replace multiple value with hte single value by giveing them in list
```

```python
df.replace(['30%','20%'],[0.0,0.0]) # replaceing multiple values with multiple values in list
```

```python
df.replace({'ID':[1,2,3,4]},333) # you can also give value in dict
```

```python
df.replace({'Industry':'health','Inception':2011},333) # multiple values with single value
```

```python
df.replace('[A-Za-z]',0,regex=True) #regex parameter take bool value
```

```python
df.replace({'Industry':'[A-Za-z]'},0,regex=True)
```

```python
df
```

```python
df.replace('Health',method='ffill')      # method repalce backfill or forward fill value with the given string
```

```python
df.replace('Health',method='bfill')
```

```python
df.replace('health',method='bfill',limit=2) # by using limit we can decide the limit of value.
```

```python
df
```

```python
df.replace('30%',55,inplace=True)
df
```
