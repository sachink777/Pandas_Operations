```python
import pandas as pd
import numpy as np
```

```python
dict1 = {'col1':[1,2,3,4,5],
        'col2':[45,56,67,34,34],
        'col3':['sachin','viraj','himanshu','sk','marri'],
    }
```

```python
dict1
```

```python
df = pd.DataFrame(dict1)
df
```

```python
df['col4'] = df['col1']**2
df
```

```python
df['col5']=df['col3'].str[0]
df
```

# apply()

```python
def sqrt(a):
    return a**2
df['col6'] = df['col1'].apply(sqrt)
df
```

```python
def sqrt(a):
    if a == 's':
        return 's'
    else:
        return 'z'
    
df['col7'] = df['col5'].apply(sqrt)
df
```

```python
df['col8'] = df['col5'].apply(lambda a : 's' if a == 's' else 'z')
df
```

```python
df['col9'] = df['col3'].apply(len)
df
```

```python
df['col10'] = df['col2'].apply(np.log10)
df
```
