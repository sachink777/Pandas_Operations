# New Dataset

Sometimes you will get dataset which may be connected with ; or any special collocation, you need to use sep parameter to
seperate out and fetch data in proper tabular format

```python
df1 = pd.read_csv(r'E:\\bank.csv', sep=";")
df1.head()
```

1. How many campaign avaliable in dataset.

```python
print(df1['campaign'].unique())
print(len(df1['campaign'].unique()), ': campaign available in dataset')
```

2. How many users do we have with housing and personal loan.

```python
df1_modify = df1[df1['housing'] == 'yes']
print(len(df1_modify[df1_modify['loan']=='yes']), ': users have housing and personal loan')
df1_modify[df1_modify['loan']=='yes']
```

3. How many person do we have whose age is 60+.

```python
print(len(df1[df1['age']>60]),' : users have age more than 60')
df1[df1['age']>60]
```

4. In which month we have targetes most of the customers.

```python
lst2 = df1['month'].unique()
for i in lst2:
    print(len(df1[df1['month']==i]),'month {} :'.format(i)) 
print('by observation we can clearly observe that in may month we have targeted most of the customers')
```

5. Which mode of call givivng you more result.

```python
lst3=df1['contact'].unique()
a=df1[df1['y']=='yes']
print(len(a[a['contact']==lst3[0]]), ': cellular result')
print(len(a[a['contact']==lst3[1]]), ': unknown result')
print(len(a[a['contact']==lst3[2]]), ': telephone result')
print('hence cellular is giving more result')
```

6. How many enterpreneurs do we have in list.

```python
print(len(df1[df1['job']== 'entrepreneur']), ': entrepreneurs we have in dataset')
df1[df1['job']== 'entrepreneur']
```

7. How many customers do we have with negative balance.

```python
print(len(df1[df1['balance']<0]), ': people we have in dataset whose balance is in the negative')
df1[df1['balance']<0]
```
