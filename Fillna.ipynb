# fillna() and its methods

loading csv dataset

```python
df=pd.read_csv("D:\dataset for Machine Learning\Fortune_10.csv")
df
```

```python
df.fillna(77)
```

we can fill Nan values of specific columns 

```python
df.fillna({'Growth':'sk'})
```

# There are few methods of filling Nan Values like ffill,pad,bfill,backfill

```python
df.fillna(method='ffill')
```

```python
df.fillna(method='pad')
```

```python
df.fillna(method='bfill')
```

```python
df.fillna(method='backfill')
```

```python
df.fillna(axis=1,method='ffill')
```

```python
df.fillna(axis=0,method='bfill')
```

we can limit the filling values by limit parameter

```python
df.fillna(value=777,limit=2)
```

```python
df.fillna(method='ffill',limit=1)
```
