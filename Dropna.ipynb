```python
import pandas as pd
```

# Dropna() and its methods

```python
df=pd.read_csv("D:\dataset for Machine Learning\Fortune_10.csv")
df
```

```python
df.dropna() # by default it drop rows
```

```python
df.dropna(axis=1) # axis=1 it drop column which having nan values
```

```python
df.dropna(how='any')     # how has by default value any
```

```python
df.dropna(axis=0,how="all")  # it drop rows or col when all the value of that r/c are null
```

```python
df.dropna(axis=1,how='any')
```

```python
df.dropna(axis=1,thresh=9) # thresh parameter is used to show minimum present value in col/row
```

```python
df.dropna(axis=1,how='any',thresh=1)
```

```python
df.dropna(subset=['Growth']) # subset  takes argument as col.name & drop the rows of the column which haveing null value.
```

```python
df.dropna(inplace=True) # inplace update same data set so if you print  dataset next time you will see some changes in original dataset
df
```
